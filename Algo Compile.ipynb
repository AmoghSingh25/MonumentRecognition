```python
def get_hog(input_img,image_name, output_dir):
    img = cv2.imread(input_img)
    img= np.float32(img) / 255.0

    gx = cv2.Sobel(img, cv2.CV_32F, 1, 0, ksize=1)
    gy = cv2.Sobel(img, cv2.CV_32F, 0, 1, ksize=1)

    mag, angle = cv2.cartToPolar(gx, gy, angleInDegrees=True)
    mag = cv2.resize(mag,(img.shape[1]//2, img.shape[0]//2))
    img_output = mag * 255.0
    img_output = img_output.astype('uint8')
    image_name = image_name[:image_name.index('.')] + '.jpg'
    cv2.imwrite(output_dir+image_name,img_output)
```

## K Means

```python
img = cv2.imread('tajmahal.jpg')
```

```python
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# reshape the image to a 2D array of pixels and 3 color values (RGB)
pixel_values = image.reshape((-1, 3))
# convert to float
pixel_values = np.float32(pixel_values)
```

```python
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
```

```python
k = 3
_, labels, (centers) = cv2.kmeans(pixel_values, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
```

```python
# convert back to 8 bit values
centers = np.uint8(centers)

# flatten the labels array
labels = labels.flatten()
```

```python
segmented_image = centers[labels.flatten()]
print(segmented_image)
```

```python
# reshape back to the original image dimension
segmented_image = segmented_image.reshape(image.shape)
# show the image
segmented_image = segmented_image.astype('uint8')
show_img(segmented_image)
```

```python
# disable only the cluster number 2 (turn the pixel into black)
masked_image = np.copy(segmented_image)
# convert to the shape of a vector of pixel values
masked_image = masked_image.reshape((-1, 3))
# color (i.e cluster) to disable
masked_image[labels == 0] = [0, 0, 0]
masked_image[labels == 1] = [0, 0, 0]
# masked_image[labels == 3] = [0, 0, 0]
# convert back to original shape
masked_image = masked_image.reshape(image.shape)
# show the image
cv2.imwrite("tajmahal_kmeans.jpg", masked_image)
masked_image = cv2.resize(masked_image, (masked_image.shape[1]//2, masked_image.shape[0]//2))
show_img(masked_image)
```

## SIFT

```python
import cv2 
import matplotlib.pyplot as plt
%matplotlib inline

# read images
img1 = cv2.imread('charminar_2.jpeg')  
img2 = cv2.imread('charminar_1.jpeg') 

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

#sift
sift = cv2.SIFT_create()

keypoints_1, descriptors_1 = sift.detectAndCompute(img1,None)
keypoints_2, descriptors_2 = sift.detectAndCompute(img2,None)

#feature matching
bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)

matches = bf.match(descriptors_1,descriptors_2)
matches = sorted(matches, key = lambda x:x.distance)

img3 = cv2.drawMatches(img1, keypoints_1, img2, keypoints_2, matches[:50], img2, flags=2)
# show_img(img3)
cv2.imwrite("image.jpg",img3)
```
